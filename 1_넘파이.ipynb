# 2. 행렬 연산

###2-1. 연산자

```python
# 덧셈 연산
a = np.array([[1, 2, 3], 
             [2, 3, 4]])
b = np.array([[3, 4, 5],
             [1, 2, 3]])
```

```python
# shape가 같아야 함
# 같은 position끼리 연산 
a + b
```

```python
a - b
```

```python
# 곱셈 연산
a = np.array([[1, 2, 3], 
             [2, 3, 4]])
b = np.array([[3, 4, 5],
             [1, 2, 3]])
```

```python
a * b
```

```python
# 나눗셈 연산
a = np.array([[1, 2, 3], 
             [2, 3, 4]])
b = np.array([[3, 4, 5],
             [1, 2, 3]])
```

```python
a / b
```

```python
# dot product(내적)
# 맞닿는 shape가 같아야 함
# 결과 행렬은 떨어져 있는 shape의 형태
a = np.array([[1, 2, 3],
              [1, 2, 3],
              [2, 3, 4]])
b = np.array([[1, 2],
              [3, 4],
              [5, 6]])
```

```python
a.shape, b.shape
```

```python
print((1*1+2*3+3*5), (1*2+2*4+3*6))
print((1*1+2*3+3*5), (1*2+2*4+3*6))
print((2*1+3*3+4*5), (2*2+3*4+4*6))
```

```python
np.dot(a, b)
```

###2-3. sort(정렬)

```python
arr1 = np.array([1, 10, 5, 7, 2, 4, 3, 6, 8, 9])
```

```python
arr1
```

```python
np.sort(arr1) # 기본적으로 오름차순 정렬
```

```python
np.sort(arr1)[::-1] # 내림차순 
```

```python
li1 = [1, 10, 5, 7, 2, 4, 3, 6, 8, 9]
```

```python
li1.sort()
```

```python
li1 # 정렬된 결과를 유지
```

```python
arr1 # 정렬된 결과를 유지하지 않음
```

```python
result = np.sort(arr1)
```

```python
result
```

```python
arr2d = np.array([[11, 10, 12, 9],
                  [3, 1, 4, 2],
                  [5, 6, 7, 8]])
```

```python
arr2d.shape
```

```python
# 행 정렬
np.sort(arr2d, axis = 0)
```

```python
# 열 정렬
np.sort(arr2d, axis = 1)
```

```python
# 축의 마지막 방향 (2차원에서는 열, 3차원에서는 다른 축방향)
np.sort(arr2d, axis = -1)
```
