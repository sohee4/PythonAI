* 특정 분석에 적합하게 데이터를 가공하는 작업
* 업무에서 사용하는 데이터는 분석, 머신러닝(딥러닝)에 적합하지 않은 경우가 많기 때문에 이를 방지하기 위한 작업을 함

# 2. 공공데이터

* [공공데이터](http://www.data.go.kr)에서 제공하는 데이터르 ㄹ활용
* [민간 아파트 가격동향](https://bit.ly/ds-house-price)

```python
import pandas as pd
```

```python
df = pd.read_csv('https://bit.ly/ds-house-price')
```

```python
df
```

```python
# 데이터 프레임의 정보확인
df.info()
```

```python
# 컬럼명 수정
df = df.rename(columns = {'분양가격(㎡)':'분양가격'})
```

```python
df
```

```python
df.info() # 분양가격이 string이라는걸 확인!
```

```python
# 분양가격의 dtype을 int64로 변경하기
# df['분양가격'].astype(int) # ValueError: invalid literal for int() with base 10: '  '
```

```python
# '  ' 공백이 존재하는 데이터가 있는지 확인
df.loc[df['분양가격'] == '  ']
```

```python
# strip()를 사용하여 공백이 있는 데이터의 공백을 삭제
df['분양가격'] = df['분양가격'].str.strip()
```

```python
df.loc[df['분양가격'] == '  ']
```

```python
# df['분양가격'].astype(int) # ValueError: invalid literal for int() with base 10: ''
```

```python
df.loc[df['분양가격'] == '']
```

```python
# 데이터가 없는 경우는 0으로 변경
df.loc[df['분양가격'] == '', '분양가격'] = 0
```

```python
df.loc[df['분양가격'] == '']
```

```python
# df['분양가격'].astype(int) # ValueError: cannot convert float NaN to integer
```

```python
# NaN을 0으로 변환
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
# df['분양가격'].astype(int) # ValueError: invalid literal for int() with base 10: '6,657'
```

```python
# 콤마를 제거
df['분양가격'] = df['분양가격'].str.replace(',','')
```

```python
# df['분양가격'].astype(int) # ValueError: cannot convert float NaN to integer
```

```python
# NaN을 0으로 변환
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
# df['분양가격'].astype(int) # ValueError: invalid literal for int() with base 10: '-'
```

```python
df['분양가격'] = df['분양가격'].str.replace('-','')
```

```python
# df['분양가격'].astype(int) # ValueError: cannot convert float NaN to integer
```

```python
# NaN을 0으로 변환
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
# df['분양가격'].astype(int) # ValueError: invalid literal for int() with base 10: ''
```

```python
# 데이터가 없는 경우는 0으로 변경
df.loc[df['분양가격'] == '', '분양가격'] = 0
```

```python
df['분양가격'].astype(int)
```

```python
df['분양가격']= df['분양가격'].astype(int)
```

```python
df.info()
```

```python
df
```

```python
# 규모구분 column에서 불필요한 '전용면적 ' 글자를 삭제
df['규모구분'] = df['규모구분'].str.replace('전용면적 ', '')
```

```python
df['규모구분']
```

```python
df['규모구분'].value_counts()
```

```python
# 지역명별로 평균 분양가격을 확인
df.groupby('지역명')['분양가격'].mean()
```

```python
# 지역별 최고 비싼 분양가
df.groupby('지역명')['분양가격'].max()
```

```python
# 연도별 분양가격의 평균
df.groupby('연도')['분양가격'].mean()
```

```python
# 작업파일 저장
df.to_csv('ds-house-price-clean.csv', index = False)
```
