#2. 스타일 옵션

```python
plt.figure(figsize = (6, 8)) # inch
plt.plot([1, 2, 3], [1, 2, 3])
plt.plot([1, 2, 3], [2, 4, 6])

plt.title('타이틀 테스트', fontsize = 20)
plt.xlabel('x축', fontsize = 15)
plt.ylabel('y축', fontsize = 15, rotation = 0)

plt.show()
```

```python
plt.figure(figsize = (15, 10))
plt.title('마커 설정', fontsize = 30)
plt.plot(np.arange(10), np.arange(10), marker = 'o', linestyle = '', color = 'b', alpha = 0.3) # alpha = 투명도
plt.plot(np.arange(10), np.arange(10)*2, marker = 'v', linestyle = '--', color = 'r', alpha = 0.5)
plt.plot(np.arange(10), np.arange(10)*3, marker = '*', markersize = 15, linestyle = '-.', color = 'y', alpha = 1)

plt.legend(['10', '10*2', '10*3'], fontsize = 15, loc = 'lower right', ncol = 3) # upper, center, lower
plt.xlim([0, 12])
plt.xticks(rotation = 30)
plt.ylim([0, 30])
plt.yticks(rotation = 90)

plt.grid()
plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Korean']
y = [60, 80, 50, 90, 20, 70]

plt.figure(figsize = (8, 5))
# plt.bar(x, y, align = 'edge', alpha = 0.7, color = 'red') # align = 'edge' -> 그래프 시작점에 xlabel 적힘
plt.bar(x, y, align = 'center', alpha = 0.7, color = 'red')
plt.ylabel('Score of Students')
plt.title('Subjects')
plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Korean']
y = [60, 80, 50, 90, 20, 70]

plt.figure(figsize = (8, 5))
plt.barh(x, y, align = 'center', color = 'pink')
# plt.barh(x, y, align = 'center', color = '#e35f62')
plt.yticks(x)
plt.xlabel('Score of Students')
plt.title('Subjects')
plt.show()
```
