# 2. KoNLPy

* 다양한 형태소 분석, 태깅 라이브러리를 파이썬에서 쉽게 사용하도록 모아놓은 라이브러리
* Hannanum: 한나눔, KAIST Sematic Web Research Center 개발
* Kkma: 꼬꼬마, 서울대학교 IDS(Intelligent Data Systems)연구실 개발
* Komoran: 코모란, Shineware에서 개발
* Mecab: 메카브, 일본어용 형태소 분석기를 한국어를 사용할 수 있도록 수정 개발
* OKT(Open Korean Text): 오픈소스 한국어 분석기, 과거 트위터 형태소 분석기

```python
from konlpy.tag import *
```

```python
# 아래 모든 클래스는 명사(nouns) 추출, 형태소(morphs) 추출, 품사(pos) 부착을 공통적으로 제공
hannanum = Hannanum()
kkma = Kkma()
komoran = Komoran()
okt = Okt()
```

```python
law[:50]
```

```python
hannanum.nouns(law[:50])
```

```python
kkma.nouns(law[:50])
```

```python
komoran.nouns('\n'.join([s for s in law[:50].split('\n') if s]))
```

```python
okt.nouns(law[:50])
```

```python
# 명사 뿐 아니라 모든 품사의 형태소를 알아내려면 morphs 라는 명령을 사용
hannanum.morphs(law[:50])
```

```python
# pos 명령을 사용하면 품사를 부착할 수 있음
# 형태소 분석기마다 사용하는 품사 태그가 다르므로 각 형태소 분석기에 대한 문서를 참조
hannanum.pos(law[:50])
```

```python
okt.pos(law[:50])
```

```python
# 부착되는 품사 태그의 기호와 의미는 tagset 속성으로 확인
okt.tagset
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
txt = '아버지가방에들어가신다'
okt.pos(txt)
```

```python
okt.pos('이런것도 되나욬ㅋㅋㅋㅋㅋㅋ')
```

```python
# norm = True : 품사 태깅(기본값 False)
okt.pos('이런것도 되나욬ㅋㅋㅋㅋ', norm = True)
```

```python
# stem = True : 원형 글자로 변경(기본값 False)
okt.pos('이런것도 되나욬ㅋㅋㅋㅋ', norm = True, stem = True)
```
